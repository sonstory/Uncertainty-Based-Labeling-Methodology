# Initial Train Data(17,295)

```python
import os
import shutil
import splitfolders
import numpy as np
import pandas as pd

data_17295 = pd.read_csv('../Active_Learning/Data/Data_17295/data_17295.csv', index_col=0)

# 디렉토리 생성
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
            
    except OSError:
        print('Error')

for num in data_17295['failureNum'].unique():
    createFolder(f'../Active_Learning/Data/Data_17295/Labeled/{num}')
```

```python
data_17295
```

```python
for idx in range(len(data_17295)):
    fail_num = data_17295.iloc[idx]['failureNum']
    image_file = str(data_17295.iloc[idx]['index']) + '.png'
    shutil.move(f'../Active_Learning/Data/Data_172950/{fail_num}/{image_file}',f'../Active_Learning/Data/Data_17295/Labeled/{fail_num}/{image_file}')
```
